```python
import os
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
#usmmarize model outputs into one dataframe 

path = '/content/drive/My Drive/INF-Msc_Thesis/Msc-Thesis-Crop-Type-Mapping/model_output/9class_mmp_5times'
model_list = os.listdir(path)
summary_df = pd.DataFrame()

for i in range(len(model_list)):
  metric_df = pd.read_csv(path + '/'+ model_list[i])
  summary_df.insert(i,model_list[i],metric_df['metrics values'])
summary_df.insert(0,'metrics',metric_df['metrics names'])
summary_df.set_index('metrics')
summary_df.head(6)

summary_df.to_csv(path + "modelperformance.csv")
```

```python
#summarize the recall of the model output to see the classification accuracy of each class

path = '/content/drive/My Drive/INF-Msc_Thesis/Msc-Thesis-Crop-Type-Mapping/model_output/9class_mmp_5times'
model_list = os.listdir(path)
summary_df = pd.DataFrame()
for i in range(len(model_list)):
  metric_df = pd.read_csv(path + '/'+ model_list[i])
  summary_df.insert(i,model_list[i],metric_df['recall'])

summary_df['mean'] = summary_df.mean(axis=1)


summary_df['mean'].to_csv(path + "meanclassrecall.csv")
```

```python
#summarize the confusion matrix of the model output to see the classification accuracy of each class prediction

path = '/content/drive/My Drive/INF-Msc_Thesis/Msc-Thesis-Crop-Type-Mapping/model_output/9class_mmp_5times'
model_list = os.listdir(path)
summary_df = np.zeros([9,9])
for i in range(len(model_list)):
  metric_df = pd.read_csv(path + '/'+ model_list[i])
  summary_df += metric_df.iloc[:,1:10].to_numpy()


summary_df = pd.DataFrame(summary_df)
summary_df = summary_df.div(summary_df.sum(axis=1), axis=0)

summary_df.to_csv(path + "meanclassconfmatrix.csv")
```
